```python
import os
import warnings
warnings.simplefilter('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread,imshow
from skimage.transform import resize      
from skimage.color import rgb2gray

```

```python
kareena=os.listdir("/content/DATASET/kareena")
```

```python
monica=os.listdir("/content/DATASET/monica")
```

```python
shakira=os.listdir("/content/DATASET/shakira")
```

```python
len(monica)
```

```python
limit = 100
monica_images = [None]*limit
j=0
for i in monica:
    if (j<limit):
        monica_images[j] = imread("/content/DATASET/monica/"+i)
        j+=1
    else:
        break
```

```python
limit = 100
shakira_images = [None]*limit
j=0
for i in shakira:
    if (j<limit):
        shakira_images[j] = imread("/content/DATASET/shakira/"+i)
        j+=1
    else:
        break
```

```python
limit = 100
kareena_images = [None]*limit
j=0
for i in kareena:
    if (j<limit):
        kareena_images[j] = imread("/content/DATASET/kareena/"+i)
        j+=1
    else:
        break
```

```python
imshow(kareena_images[22])
```

```python
imshow(shakira_images[22])
```

```python
imshow(monica_images[22])
```

```python
limit = 100
kareena_gray = [None]*limit
j = 0

for i in kareena:
    if(j<limit):
        kareena_gray[j] = rgb2gray(kareena_images[j])
        j = j+1
    else:
        break
```

```python
imshow(kareena_gray[22])
```

```python
limit = 100
monica_gray = [None]*limit
j = 0

for i in monica:
    if(j<limit):
        monica_gray[j] = rgb2gray(monica_images[j])
        j = j+1
    else:
        break
```

```python
imshow(monica_gray[22])
```

```python
limit = 100
shakira_gray = [None]*limit
j = 0

for i in shakira:
    if(j<limit):
        shakira_gray[j] = rgb2gray(shakira_images[j])
        j = j+1
    else:
        break
```

```python
imshow(shakira_gray[22])
```

```python
for j in range(100):
    kareena =  kareena_gray[j]
    kareena_gray[j]=resize(kareena,(250,250))
```

```python
imshow(kareena_gray[22])
```

```python
for j in range(100):
    monica =  monica_gray[j]
    monica_gray[j]=resize(monica,(250,250))
```

```python
imshow(monica_gray[22])
```

```python
for j in range(100):
    shakira =  shakira_gray[j]
    shakira_gray[j]=resize(shakira,(250,250))
```

```python
imshow(shakira_gray[22])
```

```python
len_of_images_kareena=len(kareena_gray)
```

```python
len_of_images_monica=len(monica_gray)
```

```python
len_of_images_shakira=len(shakira_gray)
```

```python
image_size_kareena=kareena_gray[22].shape
image_size_kareena
```

```python
image_size_monica=monica_gray[22].shape
image_size_monica
```

```python
image_size_shakira=shakira_gray[22].shape
image_size_shakira
```

```python
flatten_size_kareena=image_size_kareena[0]*image_size_kareena[1]
flatten_size_kareena
```

```python
flatten_size_monica=image_size_monica[0]*image_size_monica[1]
flatten_size_monica
```

```python
flatten_size_shakira=image_size_shakira[0]*image_size_shakira[1]
flatten_size_shakira
```

```python
for i in range(len_of_images_kareena):
    kareena_gray[i] = np.ndarray.flatten(kareena_gray[i]).reshape(flatten_size_kareena,1)
```

```python
kareena_gray = np.dstack(kareena_gray)
kareena_gray = np.rollaxis(kareena_gray, axis=2, start=0)
kareena_gray.shape
```

```python
kareena_gray = kareena_gray.reshape(len_of_images_kareena, flatten_size_kareena)
kareena_gray.shape
```

```python
for i in range(len_of_images_monica):
    monica_gray[i] = np.ndarray.flatten(monica_gray[i]).reshape(flatten_size_monica,1)
```

```python
monica_gray = np.dstack(monica_gray)
monica_gray = np.rollaxis(monica_gray, axis=2, start=0)
monica_gray.shape
```

```python
monica_gray = monica_gray.reshape(len_of_images_monica, flatten_size_monica)
monica_gray.shape
```

```python
for i in range(len_of_images_shakira):
    shakira_gray[i] = np.ndarray.flatten(shakira_gray[i]).reshape(flatten_size_shakira,1)
```

```python
shakira_gray = np.dstack(shakira_gray)
shakira_gray = np.rollaxis(shakira_gray, axis=2, start=0)
shakira_gray.shape
```

```python
shakira_gray = shakira_gray.reshape(len_of_images_shakira, flatten_size_shakira)
shakira_gray.shape
```

```python
kareenadf = pd.DataFrame(kareena_gray)
kareenadf
```

```python
kareenadf["label"] = "kareena"
kareenadf
```

```python
monicadf = pd.DataFrame(monica_gray)
monicadf
```

```python
monicadf["label"] = "monica"
monicadf
```

```python
shakiradf = pd.DataFrame(shakira_gray)
shakiradf
```

```python
shakiradf["label"] = "shakira"
shakiradf
```

```python
combine_1 = pd.concat((kareenadf,monicadf))
combine_1
```

```python
combine_2=pd.concat((combine_1,shakiradf))
combine_2
```

```python
from sklearn.utils import shuffle
```

```python
images_indexed = shuffle(combine_2).reset_index()
```

```python
images_indexed
```

```python
drop_index = images_indexed.drop(['index'],axis =1)
```

```python
drop_index
```

```python
x = drop_index.values[:,:-1]
```

```python
y = drop_index.values[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import svm
```

```python
clf = svm.SVC()
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import metrics
```

```python
accuracy = metrics.accuracy_score(y_test,y_pred)
```
